# Regression

Étant donné la valeur d'un vecteur d'entrée D-dimensionnel $ \mathbf{x} $, prédire la valeur d'une ou plusieurs variables cibles

Les modèles discutés dans cette section sont linéaires par rapport aux paramètres ajustables, pas nécessairement par rapport aux variables d'entrée.

## Créer un modèle

Dans ce cahier, notre objectif est de construire des modèles qui peuvent prédire la valeur d'une variable cible, $t$, étant donné un vecteur d'entrée, $\mathbf{x}$, où la valeur cible peut occuper n'importe quelle valeur dans un espace - bien que ici nous ne considérerons que l'espace des vecteurs réels valorisés. Nous voulons que les modèles tiennent compte de l'incertitude quant à la précision du modèle et/ou du bruit sur les données observées. Nous voulons également que le modèle fournisse des informations sur notre confiance dans une prédiction donnée.<br>

La première étape consiste à construire un modèle mathématique qui représente adéquatement les observations que nous souhaitons prédire. Le modèle que nous utiliserons est décrit dans les deux sous-sections suivantes. Il est important de noter que le modèle lui-même est indépendant de l'utilisation d'un point de vue fréquentiste ou bayésien. C'est ainsi que l'on obtient les paramètres libres du modèle qui sont affectés par l'utilisation d'approches fréquentistes ou bayésiennes. Cependant, si le modèle est un mauvais choix pour une observation particulière, alors sa capacité prédictive sera probablement médiocre que nous utilisions une approche fréquentiste ou bayésienne pour obtenir les paramètres.

Nous supposerons tout au long de ce cahier que la variable cible est décrite par

$t=y(\mathbf{x},\mathbf{w})=\epsilon$

où $ y (\mathbf{x}, \mathbf{w}) $ est une fonction encore non définie de $ \mathbf{x} $ et $ \mathbf{w} $ et $ \epsilon $ est un composant de bruit distribué .

<p style="text-align:justify">
Bruit gaussien? Les dérivations fournies ci-dessous supposent toutes un bruit gaussien sur les données cibles. Est-ce une bonne hypothèse? Dans de nombreux cas, oui. L'argument repose sur l'utilisation du Theorème central limite qui dit fondamentalement que la somme de nombreuses variables aléatoires indépendantes se comporte comme une variable aléatoire distribuée gaussienne. Le terme de bruit dans ce modèle, $ \epsilon $, peut être considéré comme la somme des caractéristiques non incluses dans la fonction de modèle, $ y(\mathbf{x}, \mathbf{w}) $. En supposant que ces caractéristiques sont elles-mêmes des variables aléatoires indépendantes, le théorème central des limites suggère qu'un modèle gaussien est approprié, en supposant qu'il existe de nombreuses caractéristiques indépendantes non comptabilisées. Il est possible qu'il n'y ait qu'un petit nombre de caractéristiques non comptabilisées ou qu'il y ait un véritable bruit non gauisien dans nos mesures d'observation, par ex. bruit de tir du capteur qui a souvent une distribution de Poisson. Dans de tels cas, l'hypothèse n'est plus valable</p>

Pour continuer, nous devons définir un modèle pour $ y (\mathbf {x}, \mathbf {w}) $. Nous utiliserons le modèle de régression linéaire général défini comme suit


$y(\mathbf{x},\mathbf{w}) = \sum_{j=0}^{M-1} w_j\phi_j(\mathbf{x}) = \mathbf{w}^T\mathbf{\phi}(\mathbf{x})$

<p style="text-align:justify">où $ \mathbf{x} $ est un vecteur d'entrée de dimension$D$, $M$ est le nombre de paramètres libres dans le modèle, $\mathbf{w}$ est un vecteur colonne des paramètres libres, et $ \phi (\mathbf{x}) = {\phi_0 (\mathbf {x}), \phi_1 (\mathbf{x}), \ldots, \phi_{M-1} (\mathbf{x})} $ with $ \phi_0 (\mathbf{x}) = 1 $ est un ensemble de fonctions de base où chaque $ \phi_i $ est dans l'espace de fonctions à valeur réelle $ {f \in \mathbf{R}^D \Rightarrow \mathbf{R} ^ 1 } $. Il est important de noter que l'ensemble des fonctions de base, $ \phi $, par rapport à $ \mathbf{x} $. De plus, notez que ce modèle définit une classe entière de modèles. Afin de construire un modèle prédictif réel pour une quantité observable, nous devrons faire une autre hypothèse sur le choix de l'ensemble des fonctions de base, $ \phi $. Cependant, aux fins de dériver des résultats généraux, nous pouvons retarder ce choix.</p>

Notez que $ \mathbf{w}^T$ est un vecteur $ 1\times M$ et que $ \mathbf{\phi} (\mathbf{x}) $ est un vecteur $ M \times 1 $ de sorte que la cible , $ y $ est un scalaire. Cela sera étendu aux variables cibles dimensionnelles $ K $ ci-dessous.

## Frequentist View: Maximum Likelihood

<p style="text-align:justify">Embarquons maintenant sur le chemin d'obtention des paramètres libres, $ \mathbf{w} $, de notre modèle. Nous commencerons à utiliser une approche fréquentiste ou du maximum de vraisemblance. Cette approche suppose que nous obtenons d'abord des données d'entraînement d'observation, $ \mathbf{t} $, et que la meilleure valeur de $ \mathbf{w} $, est celle qui maximise la fonction de vraisemblance, $ p(\mathbf{t} | \mathbf{w}) $.
Dans la condition de bruit gaussien, on peut montrer que la fonction de maximum de vraisemblance pour les données d'apprentissage est<br>

$ p(\mathbf{t} | \mathbf{X}, \mathbf{w}, \sigma^2) = \prod_{n = 1}^N ND(t_n|\mathbf{w}^T \phi(\mathbf{x}_n),\sigma^2) $<br>

$ = \frac{N}{2} \ln \frac{1} {\sigma^2} - \frac{N}{2} \ln(2 \pi) - \frac{1}{2\sigma^2}\sum_{n = 1}^N\{t_n - \mathbf{w}^T \phi(\mathbf{x}_n)\}^2 $<br>

où $ \mathbf{X} = \{\mathbf{x}_1, \ldots, \mathbf{x}_N \} $ est la valeur d'entrée définie pour les valeurs de sortie $ N $ observées correspondantes contenues dans le vecteur $ \mathbf{t} $, et $ ND ( \mu, \sigma^2) $ est la distribution normale (gaussienne). (J'ai utilisé ND au lieu de la norme N pour éviter toute confusion avec la limite de produit). En prenant le logarithme du maximum de vraisemblance et en fixant la dérivée par rapport à $ \mathbf{w} $ égale à zéro, on peut obtenir les paramètres de vraisemblance maximum donnés par les équations normales:<br>

$ \mathbf{w}_{ML} = \left (\mathbf {\Phi}^T \mathbf {\Phi} \right)^{- 1} \mathbf {\Phi}^T\mathbf{t} $<br>

où $\Phi$ est la matrice de conception $ N\times M $ avec les éléments $ \Phi_{n, j} = \phi_j (\mathbf{x}_n) $, et $ \mathbf{t} $ est le $ N \times K $ matrice des valeurs cibles de l'entraînement (pour $ K = 1 $, il s'agit simplement d'un vecteur colonne). Notez que $ \mathbf{\Phi}^T $ est une matrice $ M \times N $, de sorte que $ \mathbf{w}_{ML} = \left (\mathbf{\Phi}^T \mathbf{\Phi}\right)^{- 1} \mathbf{\Phi}^T \mathbf{t} $ est $ (M \times N) \times (N \times M) \times (M \times N) \times (N \times K) = M \times K $, où $ M $ est le nombre de paramètres libres et $ K $ est le nombre de valeurs cibles prédites pour une entrée donnée.<br>
Notez que le seul terme de la fonction de vraisemblance qui dépend de $ \mathbf{w} $ est le dernier terme.

La quantité, $ \mathbf{\Phi}^\dagger = \left (\mathbf {\Phi}^T \mathbf{\Phi} \right)^{- 1} \mathbf{\Phi}^T $ est connue comme le pseudo-inverse de Moore-Penrose de $ \Phi$. Lorsque $ \Phi^T \Phi $ est inversible, le pseudo-inverse est équivalent à l'inverse. Lorsque cette condition échoue, le pseudo-inverse peut être trouvé avec des techniques telles que la décomposition de valeurs singulières.</p>

#### Example 1

<p style="text-align:justify">
Générons des données de pour $ y = m*x+b+ \epsilon $ où $ \epsilon $ est une composante gaussienne aléatoire avec une moyenne nulle. Compte tenu de ces données, appliquons la solution du maximum de vraisemblance pour trouver des valeurs pour les paramètres $ m $ et $ b $. Étant donné que nous savons que nos données sont linéaires, nous avons choisi les fonctions de base $ \phi_0(x) = 1 $ et $ \phi_1(x) = x $. Ainsi, notre modèle sera $ y = \theta_0 \phi_0(x) + \theta_1 \phi_1(x) $, où vraisemblablement la solution devrait donner $ \theta_0 \approx b $ et $ \theta_1 \approx m $</p>

```python
import numpy as np
from matplotlib import pyplot as plt
#in order to compare between examples, set a seed in random
seed = 123456789
np.random.seed(seed)
def y(x,m,b,mu=0,sigma=1): return m*x + b + np.random.normal(mu,sigma,1)[0]
#training data, with N data points
N = 101
M = 2
t = np.empty(N)
domain_bound = 1.0/N
domain = np.empty(N)
for i in range(N): domain[i] = i*domain_bound
for i in range(N): t[i] = y(x=domain[i],m=4.89,b=0.57)
#design matrix, phi, N X M
phi = np.array([np.ones(N),domain]).T
#find the solution
#in this case case phi.T X phi is invertible so do the folloing:
temp1 = np.linalg.inv(np.dot(phi.T,phi)) #inverse of phi.T X phi
temp2 = np.dot(temp1, phi.T)
w1 = np.dot(temp2,t) #solution
print ('w1=',w1)
#assuming that phi.T X phi was not invertible we could find the pseudo inverse using the pinv function
#we expect to obtain the same solution
phi_pi = np.linalg.pinv(phi)
w2 = np.dot(phi_pi,t)
print ('w2=',w2)
#compute the model predicted values for the training data domain
predicted_t = [w2[0]+w2[1]*x for x in domain]
plt.plot(domain,t)
plt.plot(domain,predicted_t)
plt.show()
```

<p style="text-align:justify">Une idée fausse courante concernant la régression linéaire est que les fonctions de base doivent être linéaires. Ce n'est pas le cas. En effet, les fonctions de base n'ont même pas besoin d'être des polynômes. Ils doivent être linéairement indépendants, c'est-à-dire orthoganol. Seule la dépendance aux paramètres du modèle, $ \mathbf{w}_{ML} $, doit être linéaire. Notez qu'un exemple de modèle de paramètre non linéaire serait $ y = \exp(a) \sin(x) $ où $ a $ est le paramètre libre. Générons des données trigonométriques de la forme $ y = a + b \sin(x) + c \cos(x) + \epsilon $ où à nouveau $ \epsilon $ est une composante gaussienne aléatoire avec une moyenne nulle. Ici, nous avons choisi les fonctions de base $ \phi_0 = 1 $, $ \phi_1(x) = sin(x) $ et $ \phi_2(x) = cos(x)$. Si vous vous demandez si nous trichons un peu ici, la réponse est oui. En réalité, il se peut que nous ne sachions pas à l'avance quelles devraient être les fonctions de base appropriées pour les données observées. Le choix approprié peut être suggéré par les données, la connaissance du problème et d'autres techniques d'apprentissage automatique (nous espérons que cela sera discuté plus tard). Dans cet exemple, notre modèle sera $ y = \theta_0 \phi_0(x) + \theta_1 \phi_1(x) + + \theta_2 \phi_2 (x) $, où la solution devrait vraisemblablement donner $ \theta_0 \approx a $, $ \theta_1 \approx b $ et $ \theta_2 \approx c $.</p>

```python
import numpy as np
import math
from matplotlib import pyplot as plt
#in order to compare between examples, set a seed in random
seed = 123456789
np.random.seed(seed)
def y(x,a,b,c,mu=0,sigma=1): return a + b*math.sin(x) + c*math.cos(x) + np.random.normal(mu,sigma,1)[0]
#training data, with N data points
N = 101
M = 3
t = np.empty(N)
domain = np.empty(N)
domain_bound = 4.0*math.pi/N
for i in range(N): domain[i] = i * domain_bound
for i in range (N): t[i] = y(x=domain[i],a=1.85,b=0.57,c=4.37)
#design matrix, phi, N X M
c1 = [math.sin(x) for x in domain]
c2 = [math.cos(x) for x in domain]
phi = np.array([np.ones(N),c1,c2]).T
#find the solution
#in this case case phi.T X phi is invertible so do the folloing:
temp1 = np.linalg.inv(np.dot(phi.T,phi)) #inverse of phi.T X phi
temp2 = np.dot(temp1, phi.T)
w1 = np.dot(temp2,t) #solution
print ('w1=',w1)
#assuming that phi.T X phi was not invertible we could find the pseudo inverse using the pinv function
#we expect to obtain the same solution
phi_pi = np.linalg.pinv(phi)
w2 = np.dot(phi_pi,t)
print ('w2=',w2)
#compute the model predicted values for the training data domain
predicted_t = [w2[0]+w2[1]*math.sin(x)+w2[2]*math.cos(x) for x in domain]
plt.plot(domain,t)
plt.plot(domain, predicted_t)
plt.show()
```

<p style="text-align:justify">Dans les cas où l'ensemble de données d'apprentissage est très volumineux ou où les données sont reçues dans un flux, une solution directe utilisant les équations normales peut ne pas être possible. Une approche alternative est l'algorithme de descente de gradient stochastique. Si la fonction d'erreur totale, $ TE $, est la somme d'une fonction d'erreur donnée, $ E $, évaluée à chacune des entrées de test $ N $, $ TE = \sum_{i = 1}^NE (\mathbf{ x}_i) $ alors l'algorithme de descente de gradient stochastique est<br>

$ \mathbf{w}^{\tau + 1} = \mathbf{w}^\tau- \eta \bigtriangledown E_\tau $<br>

où $ {\tau} $ est le numéro de l'itération et $ \eta $ est un paramètre de taux d'apprentissage. Pour ce type de fonction d'erreur totale, l'ordre d'évaluation ne change pas le résultat. Si la fonction d'erreur est la fonction somme des carrés, alors l'algorithme est

$ \mathbf{w}^{\tau + 1} = \mathbf{w}^\tau + \eta \left(t_n - \mathbf{w}^{(\tau)T} \phi_n\right) \phi_n $</p>

#### Example 2

Répétez l'évaluation de l'exemple 1(b) en utilisant l'approche de descente de gradient. La réponse est-elle la même lorsque vous utilisez les mêmes données d'entraînement?

```python
import numpy as np
import math
from matplotlib import pyplot as plt
#in order to compare between examples, set a seed in random
seed = 123456789
np.random.seed(seed)
def y(x,a,b,c,mu=0,sigma=1): return a + b*math.sin(x) + c*math.cos(x) + np.random.normal(mu,sigma,1)[0]
N = 101
M = 3
w = np.zeros((M,1))
phi = np.empty((M,1))
eta = 0.25
#create arrays to store the values as they are generated so they can be plotted at the end
x = np.empty(N)
t = np.empty(N)
domain_bound = 4.0*math.pi/N
for i in range(N):
    x[i] = i * domain_bound
    t[i] = y(x[i],a=1.85,b=0.57,c=4.37)
    phi = np.array([[1],[math.sin(x[i])],[math.cos(x[i])]]) 
    w = w + eta*(t[i]-np.dot(w.T,phi))*phi #the learning model

print(w.T)
#compute the model predicted values for the training data domain
predicted_t = [w[0]+w[1]*math.sin(item)+w[2]*math.cos(item) for item in x]
plt.plot(x,t)
plt.plot(x,predicted_t)
plt.show()
```

### Over and Under Fitting

Cette section décrit certaines des techniques utilisées pour tenir compte du sur-ajustement ou du sous-ajustement d'un modèle donné lors de l'utilisation d'une approche du maximum de vraisemblance.

#### Regularized Least Squares

<p style="text-align:justify">La régularisation tente de résoudre le problème de surajustement d'un modèle donné aux données d'apprentissage et/ou d'éliminer les paramètres qui ne sont pas importants sur la base des données. La fonction d'erreur totale générale avec un terme de régularisation est donnée par

$ E_D(\mathbf{w}) + \lambda E_W(\mathbf{w}) $

où $ \lambda $ est le coefficient de régularisation et $ E_W $ le terme de régularisation. Un terme de régularisation couramment utilisé est la somme des carrés des éléments de paramètres du modèle

$ E_W (\mathbf{w}) = \frac1{2} \mathbf{w}^T \mathbf{w} $

connu sous le nom de régularisateur de décroissance du poids. Ces termes de régularisation conduisent à la solution optimale, en supposant un modèle de régression linéaire avec bruit gaussien sur les données d'apprentissage, de

$ \mathbf{w} = \left (\lambda \mathbf{I} + \mathbf {\Phi}^T \mathbf {\Phi} \right)^{-1} \mathbf {\Phi}^T \mathbf{t} $ </p>

#### Example 3

<p style="text-align:justify">Dans cet exemple, nous utilisons les mêmes données d'apprentissage que l'exemple 1 (a), sauf qu'ici, le modèle est choisi par erreur pour être un polynôme d'ordre 7. Cet exemple est quelque peu artificiel, mais il illustre le fait qu'un modèle de surajustement peut être corrigé dans une certaine mesure en utilisant la régularisation.</p>

```python
import numpy as np
from matplotlib import pyplot as plt
#in order to compare between examples, set a seed in random
seed = 123456789
np.random.seed(seed)
def y(x,m,b,mu=0,sigma=1): return m*x + b + np.random.normal(mu,sigma,1)[0]
def el_pow(x,pow):
    temp = x
    for i in range(pow-1):
        temp = temp * x
    return temp
def prediction(params, x):
    pred = 0
    for i in range(len(params)):pred += params[i]*math.pow(x,i)
    return pred
#training data, with N data points
N = 101
M = 8
t = np.empty(N)
domain = np.empty(N)
domain_bound = 1.0/N
for i in range(N): domain[i] = i*domain_bound
for i in range(N): t[i] = y(x=domain[i],m=4.89,b=0.57)
#find the solution without using regularization
#design matrix, phi, N X M
phi = np.array([np.ones(N),domain, el_pow(domain,2),el_pow(domain,3),el_pow(domain,4),el_pow(domain,5),el_pow(domain,6),el_pow(domain,7)]).T
temp1 = np.linalg.inv(np.dot(phi.T,phi)) #inverse of phi.T X phi
temp2 = np.dot(temp1, phi.T)
w1 = np.dot(temp2,t) #solution
print('w1=',w1)
predicted_t = [prediction(w1,x) for x in domain]

#find the regularized solution
lam = 0.1
temp1 = np.linalg.inv(lam*np.eye(M)+np.dot(phi.T,phi))
temp2 = np.dot(temp1,phi.T)
w2 = np.dot(temp2,t)
print('w2=',w2)
predicted_t_reg = [prediction(w2,x) for x in domain]

#add some plots
plt.plot(domain,t)
plt.plot(domain,predicted_t)
plt.plot(domain,predicted_t_reg)
plt.legend(("training","un-regularized","regularized"), loc='lower right')
plt.show()
```

### Locally Weighted Linear Regression

En essayant de prédire une valeur cible, $ y $, pour une nouvelle entrée $ x $ et étant donné un ensemble d'apprentissage $ \mathbf{t} $, les poids de régression linéaire pondérés localement entraînent les entrées par rapport à leur distance de la nouvelle entrée $ x $ . Le modèle de régression ajusté consiste à trouver $ \theta $ pour minimiser <br>

$ \sum_i w^{(i)} \left(y^{(i)} - \theta^Tx^{(i)} \right) $ <br>

où<br>

$ w^{(i)} = \exp \left (- \frac {\left (x^{(i)} - x \right)^2} {\left (2 \tau^2 \right)} \right) $<br>

Le paramètre $ \tau $, appelé bande passante, contrôle la façon dont le poids d'une entrée d'entraînement, $ x^{(i)} $, diminue avec la distance de la nouvelle entrée, $ x $. Notez également que les poids, $ w^{(i)} $, nécessitent que les données d'entraînement soient disponibles pour chaque évaluation d'une nouvelle entrée $ x $, ce qui en fait un soi-disant modèle.<br>

La solution est obtenue comme<br>

$ \mathbf{w}_{ML} = \left (\mathbf {\Phi}^T \mathbf{W} \mathbf {\Phi} \right)^{- 1} \mathbf {\Phi}^T \mathbf{W} \mathbf{t} $

où $ \mathbf{W} $ est la matrice de pondération $ N \times N $ avec les éléments $ w_{i, j} = \left [w^{(i)} \right]^2 $ pour $ i = j $ et 0 sinon. </p>

#### Example 4

Here we apply locally weighted regression to training data that is generated from a second order polynomial with Gaussian noise.

```python
from matplotlib import pyplot as plt
#in order to compare between examples, set a seed in random
seed = 123456789
np.random.seed(seed)
def y(x,coefs,inc_ran = True,mu=0,sigma=0.2): 
    ans = 0
    for i in range(len(coefs)): ans += coefs[i]*math.pow(x,i)
    if inc_ran: return ans + np.random.normal(mu,sigma,1)[0]
    else: return ans
#training data, with N = 101 data points
N = 101
M = 2
t = np.empty(N)
domain = np.empty(N)
domain_bound = 3.0/N
for i in range(N): domain[i] = i*domain_bound
for i in range(N): t[i] = y(x=domain[i],coefs=[1.75, 0.25, -1.0])

#first find the standard linear regression fit using model y = phi_0 + phi_1 * x
#design matrix, phi, N X M
phi = np.array([np.ones(N),domain]).T
temp1 = np.linalg.inv(np.dot(phi.T,phi)) #inverse of phi.T X phi
temp2 = np.dot(temp1, phi.T)
w1 = np.dot(temp2,t) #solution
predicted_t = [y(x,w1,inc_ran=False) for x in domain]

#now construct the locally weighted solution
def wt(x,x_i,tau=1): return math.exp(-(x_i-x)*(x_i-x)/(2*tau*tau))
def lws(x, wt_func, td_x,td_y, Phi):
    _M = Phi.shape[1]
    _N = Phi.shape[0]
    _W = np.zeros((_N,_N))
    for i in range(_N): _W[i,i] = wt_func(x,td_x[i])
    temp1 = np.linalg.inv(np.dot(np.dot(Phi.T,_W),Phi)) #inverse of phi.T X phi
    temp2 = np.dot(np.dot(temp1,Phi.T),_W)
    w1 = np.dot(temp2,t) #local solution for parameters
    return y(x,w1,inc_ran=False)
predicted_local = [lws(x, wt, domain, t, phi) for x in domain]  
plt.plot(domain,t)
plt.plot(domain,predicted_t)
plt.plot(domain,predicted_local)
plt.legend(('training','regression', 'local regression'),loc='lower left')
```

### Vue bayésienne

Nous considérons ici une approche entièrement bayésienne du problème de régression. Notez tout d'abord que nos hypothèses fondamentales de modèle sont toujours valables. À savoir, nous supposons que l'observation que nous essayons de prédire est modélisée par

$ t = y (\mathbf {x}, \mathbf {w}) + \epsilon $

où $ \epsilon $ est une composante de bruit distribuée gaussienne. De plus, l'hypothèse d'une dépendance linéaire sur les paramètres du modèle, $ \mathbf {w} $ est également valable:

$ y (\mathbf{x}, \mathbf{w}) = \sum_{j = 0}^{M-1}w_j\phi_j(\mathbf{x}) = \mathbf{w}^T \mathbf{ \phi}(\mathbf{x}) $

Rappelons que l'approche bayésienne implique que les meilleurs paramètres du modèle, $ \mathbf{w} $, sont ceux qui maximisent la probabilité postérieure, qui, d'après le théorème de Bayes, est donnée par:

$ p (\mathbf{w} | \mathbf{t}) = \frac {p (\mathbf{t} | \mathbf{w}) p (\mathbf{w})}{p (\mathbf{t} )} $

#### Modèle antérieur

La première étape de la formulation d'un modèle bayésien consiste à construire un modèle pour le modèle de probabilité antérieur, $ p (\ mathbf {w}) $. REMARQUE: En général, tout modèle de distribution de probabilité peut être choisi pour modéliser le précédent. Le choix approprié dépend de différents facteurs, y compris la connaissance préalable du problème et la commodité mathématique. Souvent, le prior est choisi pour être le prior conjugué de la fonction de vraisemblance. C'est un choix de convenance mathématique car il implique que le postérieur peut être dérivé analytiquement. C'est souvent un choix raisonnable, comme c'est le cas ici pour le problème de la régression linéaire. Ainsi, étant donné la distribution gaussienne utilisée pour la fonction de vraisemblance ci-dessus, nous supposons une distribution gaussienne pour notre a priori de la forme:

$ p (\mathbf{w}) = ND (\mathbf{w} | \mathbf{m} _0, \mathbf{S}_0) $

où $ \mathbf{m}_0 $ est la moyenne antérieure et $ \mathbf{S}_0 $ est la covariance antérieure. Nous supposerons $ \mathbf{m}_0 = 0 $ et un avant infiniment large de sorte que $ \mathbf{S}_0 = \alpha^{- 1} \mathbf{I} $ où $ \alpha $ est un paramètre de précision que nous devrons choisir. Compte tenu de ces choix de fonctions a priori et de vraisemblance, la probabilité a posteriori est donnée par

$ p (\mathbf{w} | \mathbf{t}) = N \left (\mathbf {w} | \mathbf{m}_N, \mathbf{S}_N \right) $

où

$ \mathbf{m}_N = \beta \mathbf{S}_N \mathbf {\Phi}^T \mathbf{t} $

$ \mathbf{S}_N^{- 1} = \alpha \mathbf{I} + \beta \mathbf{\Phi}^T \mathbf {\Phi} $

où $ \beta = 1 / \sigma^2 $ est la variance inverse de la composante de bruit aléatoire associée à la variable cible, $ t = y (\mathbf{x}, \mathbf{w}) + ND (0, \sigma^2) $. Enfin, le log du postérieur est considéré comme la somme du log de vraisemblance et du log du a priori:

$ \ln p (\mathbf{w} | \mathbf{t}) = - \frac{\beta}{2} \left[\sum_{n = 1}^N {t_n - \mathbf {w}^T \phi (\mathbf {x} _n)}^2 \right] - \frac {\alpha}{2} \mathbf{w}^T \mathbf{w} + constante $

#### NOTE:

Plutôt que de trouver une estimation ponctuelle pour $ \mathbf {w} $ en maximisant le postérieur et ainsi faire des prédictions ponctuelles pour la variable cible $ t $, il est plus instructif d'utiliser le postérieur pour formuler une distribution prédictive pour $ t $. Pour nos hypothèses de modèle, cela est donné par

$ p (t | \mathbf {t}, \alpha, \beta) = \int p (t | \mathbf {w}, \beta) p (\mathbf {w} | \mathbf {t}, \alpha, \beta) d \mathbf{w} = ND (t | \mathbf{m}_N^T \phi (\mathbf{x}), \frac{1}{\beta} + \phi(\mathbf{x})^T \mathbf{S}_N \phi(\mathbf{x}) $

où une estimation ponctuelle de $ t $ est donnée ma moyenne $ \mu = \mathbf{m}_N^T \phi (\mathbf{x}) $ et une estimation de l'incertitude est donnée par l'écart type $ \sigma_N^2 (\mathbf{x}) = \frac{1}{\beta} + \phi(\mathbf{x})^T \mathbf{S}_N \phi (\mathbf{x}) $

Il y a un dernier problème pour compléter le modèle bayésien, à savoir la détermination de $ \alpha $ et $ \beta $. Cela peut être fait de manière entièrement bayésienne en développant des modèles antérieurs, mais cela tend à rendre les équations insolubles. Au lieu de cela, l'approche dite de la fonction de preuve est utilisée. Voir la section 3.5 de Bishop pour plus de détails.

#### Example 5

Utilisez les données fournies dans l'exemple 5 pour formuler un modèle bayésien. Ici, nous savons que $ \beta = (1 / 0,2) = 5 $. Pour cet exemple, nous choisirons simplement $ \alpha = .4 $ de sorte que $ \alpha / \beta = 0.08 $ (voir Bishop 3.5 pour les méthodes de détermination de $ \alpha $ et $ \beta $.

```python
import numpy as np
from matplotlib import pyplot as plt
#in order to compare between examples, set a seed in random
seed = 123456789
np.random.seed(seed)
alpha = 0.4
beta = 5.0
def y(x,coefs, mu=0, sigma=1.0/beta): 
    ans = 0
    for i in range(len(coefs)): ans += coefs[i]*math.pow(x,i)
    return ans + np.random.normal(mu,sigma,1)[0]
    
#training data, with N = 101 data points
N = 101
M = 4
t = np.empty(N)
domain = np.empty(N)
domain_bound = 3.0/N
for i in range(N): domain[i] = i*domain_bound
for i in range(N): t[i] = y(x=domain[i],coefs=[1.75, 0.25, -1.0])

#Let's assume that we want to fit a 3rd order polynomial to the data even though we know its a second order
#polynomial. Given the Bayesain approach, we should see the so that the unecessary terms are damped out. We have 
#y = phi_0 + phi_1 * x + phi_2 x^2 + phi_3 x^4
#design matrix, phi, N X M where N = 101 and M = 4
d2 = domain * domain
phi = np.array([np.ones(N),domain, d2, d2 * domain]).T
alphaI = alpha * np.eye(M)
SN = np.linalg.inv(alphaI + beta * np.dot(phi.T,phi)) #posterior variance
mN = beta * np.dot(np.dot(SN, phi.T), t)
point_estimates = [np.dot(mN, phi[i]) for i in range(N)]
uncertain_t = [1.0/beta + np.dot(np.dot(phi[i].T, SN), phi[i]) for i in range(N)]
plt.plot(domain,t)
plt.errorbar(domain,point_estimates, uncertain_t, ecolor = "red")
plt.legend(('training','Bayes'),loc='lower left')
```
