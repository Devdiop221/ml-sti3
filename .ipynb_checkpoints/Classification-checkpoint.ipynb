# Introduction

<p style="text-align:justify">Ce cahier décrit plusieurs modèles linéaires de classification, c'est-à-dire étant donné un vecteur d'entrée $ \mathbf{x} $ déterminer $ p(C_k | \mathbf{x}) $ où $ k \in  {1 \ldots K } $ et $ C_k $ est une étiquette de classe discrète. Notez que les algorithmes décrits ici peuvent également être appliqués à l'entrée transformée, $ \phi(x) $, pour déterminer $ p(C_k | \phi (\mathbf{x})) $, où $ \phi $ peut être un non-linéaire transformation de l'espace d'entrée.</p>
<p style="text-align:justify">Notre première hypothèse de modèle est que notre sortie cible peut être modélisée comme</p>

$ y (\phi (\mathbf{x})) = f (\mathbf{w}^T\phi(\mathbf{x})) $

<p style="text-align:justify">où $ y $ sera un vecteur de probabilités avec des éléments $ K $. Les éléments de $ y $ sont $ y_i = p(C_i| \phi(\mathbf{x})) $ c'est-à-dire la probabilité que l'étiquette de classe correcte soit $ C_i $ étant donné le vecteur d'entrée $ \mathbf{x} $. Souvent, nous aurons simplement que $ \mathbf{w}^T \phi (\mathbf{x}) = \mathbf{w}^T \mathbf{x} + w_0 $ auquel cas nous omettons simplement $ \phi $ de la notation. La fonction $ f $ est connue sous le nom de fonction d'activation et son inverse est connu sous le nom de fonction de lien. Notez que $ f $ sera souvent non linéaire.</p>

<p style="text-align:justify">Il convient de noter que presque tout le matériel présenté ne constitue pas un traitement entièrement bayésien du problème de classification. Ceci est principalement dû au fait qu'une approche bayésienne du problème de classification est mathématiquement insoluble. Cependant, le théorème de Bayes apparaîtra souvent dans la discussion, ce qui peut prêter à confusion. Le cas échéant, j'essaierai de clarifier la différence.
</p>
<p style="text-align:justify">Un autre facteur à garder à l'esprit est que la variable cible, $ y $, décrite dans notre modèle ci-dessus, ne fournit pas de décision pour une affectation de classe pour une entrée donnée $ \mathbf{x} $. Dans les cas du monde réel où il est nécessaire de prendre une décision quant à la classe $ \mathbf{x} $ à attribuer, il faut appliquer une étape de modélisation supplémentaire basée sur la théorie de la décision. Il existe une variété de modèles de décision, qui utilisent tous les modèles de probabilité postérieure de classe, $ p (C_k | \mathbf{x}) $, tels que</p>

* Minimiser le taux d'erreurs de classification - cela correspond effectivement à choisir la classe avec la probabilité la plus élevée pour un $ \mathbf{x} $ donné
* Minimiser la perte attendue - minimise la valeur attendue d'une fonction de perte donnée, $ L $, sous la distribution de probabilité postérieure de classe
* Option de rejet

<p style="text-align:justify">Les modèles présentés ici sont appelés linéaires car, lorsque le critère de décision est celui de minimiser le taux d'erreurs de classification, ils divisent l'espace d'entrée en régions $ K $, où les frontières entre régions sont des fonctions linéaires du vecteur d'entrée $ \mathbf{x} $. Les limites de décision correspondront à où $ \mathbf{w}^T \mathbf{x} = constante $, et représenteront donc une fonction linéaire de $ \mathbf 
{x} $. Dans le cas d'une entrée transformée, $ \phi (\mathbf{x}) $, les limites de décision correspondront à où $ \mathbf{w}^T \phi (\mathbf{x}) = constant $, et représenteront donc une fonction linéaire de $ \phi (\mathbf{x}) $.</p>

# Modèles génératifs probabilistes

<p style="text-align:justify">Notre première étape consiste à former un modèle des probabilités postérieures de classe pour l'étape d'inférence. Cela peut être fait directement, comme dans la modélisation discriminante probabiliste (voir section suivante), ou en modélisant d'abord les densités conditionnelles de classe, $ p (\mathbf{x}| C_k $), et les a priori de classe, $ p(C_k) $, puis appliquer le théorème de Bayes pour obtenir le modèle postérieur. Dans cette section, nous appliquerons cette dernière approche. Nous verrons que notre modèle de probabilité a posteriori de classe dépendra uniquement de l'entrée $ \mathbf{x} $ (ou d'une transformation fixe utilisant les fonctions de base $ \phi $) et des étiquettes de classe, $ C_k $. Dans ce cas, pour les classes $ K $, notre fonction d'activation prendra la forme</p>
$$ f = p (C_k | \mathbf {x}) = \frac {\exp (a_k)} {\sum_j \exp (a_j)} $$


<p style="text-align:justify">connue sous le nom de fonction softmax où $ a_j = \ln (p (\mathbf{x} | C_k)p(C_k)) $. Dans le cas de $ K = 2 $, la fonction d'activation se réduira à la fonction sigmoïde logistique</p>

$$ f = p(C_1| \mathbf {x}) = \frac{1}{1+ \exp(-a)} = \sigma(a) $$

où $ a = \ln \frac{p (\mathbf{x} | C_1) p(C_1)} {p(\mathbf{x} | C_2) p(C_2)} $

<p style="text-align:justify">Notez que ce n'est pas la seule forme possible pour les modèles de classe postérieure. Par exemple, on peut également ajouter un terme de bruit pour tenir compte des erreurs de classification. Cette forme particulière de la fonction d'activation est une conséquence du modèle que nous choisissons pour $ p (\mathbf{x} | C_k) $ dans les sections ci-dessous. Montrer le résultat peut être un peu «mettre la charrue avant les boeufs», mais cela simplifiera la notation au fur et à mesure que nous procéderons.</p>

<p style="text-align:justify">Notez que bien que nous ayons appliqué le théorème de Bayes, ce n'est pas un modèle bayésien. Nulle part nous n'avons modélisé le paramètre probabilité postérieure $ p (\mathbf{w} | \mathbf{t}) $. En effet, nous verrons bientôt que nous utiliserons une approche du maximum de vraisemblance pour déterminer $ \mathbf{w} $.</p>

<p style="text-align:justify">Ces modèles sont appelés modèles génératifs car ils peuvent être utilisés pour générer des données d'entrée synthétiques en appliquant l'échantillonnage par transformation inverse à la distribution marginale pour $ \mathbf{x} $ définie par</p>

$ p (\mathbf{x}) = \sum_k p (\mathbf{x} | C_k) p(C_k) $

<p style="text-align:justify">Pour aller de l'avant, il est nécessaire de commencer à faire des hypothèses de modèle. Ici, nous supposerons que nous avons des entrées continues, c'est-à-dire $ x \in \Re $ (voir Bishop 202 & Ng Lecture 5 pdf pour les entrées discrètes utilisant Naive Bayes & Laplace), et que les densités conditionnelles de classe, $ p (\mathbf{x} | C_k) $ sont modélisés par une distribution gaussienne. Sous l'hypothèse gaussienne, la densité conditionnelle de classe pour la classe $ C_k $ est</p>


$ p (\mathbf {x} | C_k) = \frac {1} {\left (2 \pi \right)^{D / 2}} \frac{1}{|\mathbf{\Sigma}|^{ 1/2}} \exp{-\frac{1}{2} (\mathbf{x} - \mathbf{\mu}_k)^T \mathbf{\Sigma}^{- 1} (\mathbf{x}- \mathbf{\mu}_k) } $


<p style="text-align:justify">où $ D $ est la dimension du vecteur d'entrée $ \mathbf{x} $, $ \mathbf {\Sigma} $ est la matrice de covariance et $ \mathbf {\mu} $ est le vecteur moyen et où nous avons supposé que toutes les classes partagent la même matrice de covariance.</p>

Dans le cas de deux classes, ce résultat est substitué dans la fonction sigmoïde logistique et se réduit à


$ p (C_1 | \mathbf {x}) = \sigma \left ( \mathbf{w}^T \mathbf{x} + w_0\right) $


avons-nous défini


$ \mathbf{w} = \mathbf{\Sigma}^{- 1} \left (\mathbf {\mu}_1 - \mathbf {\mu}_2 \right) $


$ w_0 = - \frac{1}{2} \mathbf{\mu}_1^T \mathbf{\Sigma}^{- 1} \mathbf{\mu}_1 + \frac{1}{2} \mathbf{\mu}_2^T \mathbf{\Sigma}^{- 1} \mathbf{\mu}_2 + \ln \frac{p (C_1)}{p (C_2)} $


<p style="text-align:justify">Notez que les probabilités a priori de classe, $ p (C_k) $, agissent effectivement comme un terme de biais. Notez également que nous n'avons pas encore spécifié de modèle pour ces distributions. Si nous devons utiliser le résultat ci-dessus, nous devrons faire une autre hypothèse de modèle. Ici nous supposerons que les a priori de classe sont modélisés par une distribution de Bernoulli avec $ p (C_1) = \gamma $ et $ p (C_2) = 1- \gamma $.</p>

Ces résultats peuvent être étendus au cas des classes $K\gt 2$ pour lesquelles on obtient


$ a_k (\mathbf {x}) = \left [\mathbf {\Sigma}^{- 1} \mathbf{\mu}_k \right]^T \mathbf{x} - \frac{1}{2} \mathbf{\mu}_k^T \mathbf{\Sigma}^{- 1} \mathbf{\mu}_k + \ln p(C_k) $


qui est utilisé dans la première fonction d'activation fournie au début de cette section.

<p style="text-align:justify">Notez que nous n'avons toujours pas formulé de modèle complet pour les densités de classes postérieures, dans la mesure où nous n'avons pas encore résolu pour les paramètres du modèle, $ \mathbf{\mu} $ et $ \mathbf{\Sigma} $. Nous faisons cela maintenant en utilisant une approche du maximum de vraisemblance.</p>

## Maximum Likelihood Solution

<p style="text-align:justify">Considérant le cas de deux classes, $ C_1 $ et $ C_2 $, avec les distributions antérieures de Bernoulli, $ p (C_1) = \pi $ et $ p (C_2) = 1- \pi $, et avec les distributions de densité conditionnelles de classe gaussiennes $ p (\mathbf{x} | C_k) $, supposons que nous ayons un ensemble de données d'apprentissage, $ \mathbf{X} $, avec $ N $ éléments de la forme $  {\mathbf{x}_n,t_n} $ où $ t_n = 0 $ indique que $ \mathbf{x_n} $ est dans la classe $ C_1 $ et $ t_n = 1 $ indique que $ \mathbf{x_n} $ est dans la classe $ C_2 $. La fonction de vraisemblance est alors donnée par</p>
$$ p \left (\mathbf{t}, \mathbf{X} \mid \pi, \mathbf {\mu}_1, \mu_2, \mathbf{\Sigma} \right) = \prod_{n = 1}^N \left [\pi ND \left (\mathbf{x}_n \mid \mu_1, \mathbf{\Sigma} \right) \right]^{t_n} \left [\left (1- \pi \right ) ND \left (\mathbf{x}_n \mid \mu_2, \mathbf{\Sigma} \right) \right]^{1-t_n} $$


<p style="text-align:justify">En prenant la dérivée de cette expression par rapport aux différents paramètres du modèle, $ \pi $, $ \mu_1 $, $ \mu_2 $ et $ \mathbf{\Sigma} $, et en la définissant égale à zéro, nous obtenons ce qui suit estimations<:P>


$ \pi = \frac{N_1}{N_1 + N_2} $

où $ N_1 $ est le nombre d'entrées d'entraînement dans la classe $ C_1 $ et $ N_2 $ est le nombre dans la classe $ C_2 $.


$ \mu_1 = \frac{1}{N_1} \sum_{n = 1}^N t_n \mathbf{x_n} $


$ \mu_2 = \frac{1}{N_2} \sum_{n = 1}^N(1-t_n) \mathbf{x_n} $


$ \mathbf{\Sigma} = \frac{1}{N} \left [\sum_{n \in C_1} (\mathbf{x_n} - \mu_1) (\mathbf{x_n} - \mu_1)^T + \sum_{n \in C_2} (\mathbf {x_n} - \mu_2) (\mathbf{x_n} - \mu_2)^T \right] $

#### Example 1

<p style="text-align:justify">Nous choisirons des valeurs de vérité pour nos paramètres et utiliserons notre modèle génératif pour générer des données synthétiques. Nous pouvons ensuite utiliser ces données comme entrée de la solution de vraisemblance maximale pour voir les estimations des paramètres de vérité. Utilisons l'entrée 1-D pour plus de simplicité. Nous utiliserons une forme basique d'échantillonnage par transformée inverse. Plus précisément, nous souhaitons que nos données d'entrée d'entraînement, $ \mathbf{x} $, soient dérivées d'une distribution modélisée par la distribution marginale $ p (\mathbf{x}) $. Pour obtenir cela, nous formulons d'abord la fonction de distribution cumulative, $ CDF (\mathbf{X}) = \int_{- \infty}^{\mathbf{X}}p(\mathbf{x})d\mathbf{x} $. Notez que la plage du CDF est $ [0,1] $. Pour obtenir des valeurs $ \mathbf{x} $ correctement distribuées, nous choisissons une valeur d'une distribution uniforme sur $ [0,1] $, disons $ y $, et trouvons la valeur $ \mathbf{X} $ telle que $ CDF (\mathbf{X}) = y $. Cette valeur de $ \mathbf {X} $ est notre entrée $ \mathbf{x} $. Cette approche nous oblige à trouver l'inverse du CDF, ce que nous ferons numériquement. Une fois que nous avons obtenu une valeur pour $ \mathbf{x} $, nous devons l'affecter à une classe particulière. Nous supposerons que la classe correcte est celle pour laquelle la probabilité postérieure $ p (C_k | \mathbf{x}) $ est la plus grande, à moins que la différence entre les deux probabilités postérieures soit inférieure à une valeur minimale, auquel cas nous choisirons au hasard - cela ajoutera du «bruit» à nos données d'entraînement d'entrée.</p>

```python
import numpy as np
import math
from functools import reduce
from matplotlib import pyplot as plt
import scipy.integrate as sciInt
import scipy.optimize as sciOpt

#select truth data values, these will NOT be known to the training algorithm, they
#are only used in generating the sample data
tu1 =-2.0
tu2 = 2.0
tSigma = 2.0
tGamma = 0.5

#the class conditional probability p(x|Ck)
def p_xCk(x, mu, sigma):
    denom = math.sqrt(2.0 * math.pi * sigma)
    arg = -0.5 * (x - mu) * (x - mu) / sigma
    return math.exp(arg) / denom

#the marginal probability p(x)
def p_x(x, mu1, mu2, sigma, gamma):
    return gamma * p_xCk(x, mu1, sigma) + (1.0 - gamma) * p_xCk(x, mu2, sigma)

#posterior class probability vector (p(C_1|x), p(C_2|x))
def p_Ckx(x, mu1, mu2, sigma, gamma):
    a = math.log(p_xCk(x, mu1, sigma)*gamma/(p_xCk(x,mu2,sigma)*(1-gamma)))
    pc1 = 1.0/(1.0 + math.exp(-a))
    return (pc1, 1.0 - pc1)

domain = range(0,100,1)
domain = [(-5.0 + x/10.0) for x in domain]
f, axarr = plt.subplots(2,2)
f.subplots_adjust(right=1.5)
f.subplots_adjust(top=1.5)

#plot p(x|Ck)
pxC1 = [p_xCk(x,tu1,tSigma) for x in domain]
pxC2 = [p_xCk(x,tu2,tSigma) for x in domain]
ax1 = axarr[0,0]
ax1.plot(domain, pxC1)
ax1.plot(domain, pxC2)
ax1.set_ylabel('p(x|C_k)')

#plot the marginal distribution p(x)
px = [p_x(x, tu1, tu2, tSigma, tGamma) for x in domain]
ax2 = axarr[0,1]
ax2.plot(domain,px)
ax2.set_ylabel('p(x)')

#plot the posterior distributions p(C_1|x) & p(C_2|x)
pc1x = []
pc2x = []
for x in domain:
    pck = p_Ckx(x, tu1, tu2, tSigma, tGamma)
    pc1x.append(pck[0])
    pc2x.append(pck[1])
ax3 = axarr[1,0]
ax3.plot(domain, pc1x)
ax3.plot(domain, pc2x)
ax3.set_ylabel('p(C_k|x)')

#the cumulative distribution function P(x<X)
def cdf(x, mu1, mu2, sigma, gamma):
    return sciInt.quad(func=p_x, a=-np.inf, b=x, args=(mu1, mu2, sigma, gamma))

#inverse of the CDF
def inv_cdf(y, mu1, mu2, sigma, gamma):
    def f(x):
        return cdf(x,mu1,mu2,sigma,gamma)[0] - y
    return sciOpt.newton(f, 0)

#this code can be used to check our inv_cdf func, if its correct, domain and cdfs should be the same
#domain = [x/10.0 for x in range(12)[1:11]]
#icdfs = [inv_cdf(x, tu1, tu2, tSigma, tGamma) for x in domain]
#cdfs = [cdf(x, tu1, tu2, tSigma, tGamma)[0] for x in icdfs]
#print domain
#print icdfs
#print cdfs

seed = 123456789
np.random.seed(seed)
num_samples = 100
x = np.zeros(num_samples)
t = np.zeros(num_samples)
pcx = np.zeros(num_samples)
n1 = 0
nae = 0
assignment_epsilon = 0.5
#assign x to 1 for C1 and 0 for C2
for i in range(num_samples):
    rv = np.random.uniform()
    x[i] = inv_cdf(rv, tu1, tu2, tSigma, tGamma)
    pcx1 = p_Ckx(x[i], tu1, tu2, tSigma, tGamma)
    pcx2 = pcx1[1]
    pcx1 = pcx1[0]
    #we don't want a perfect dividing line for our domain, otherwise why would we need a learning algorithim?
    if math.fabs(pcx2-pcx1) <= assignment_epsilon:
        nae = nae + 1
        if np.random.uniform() <= 0.5:
            t[i] = 1
            n1 = n1 + 1
        else:
            t[i] = 0
    elif pcx1 > pcx2:
        t[i] = 1
        n1 = n1 + 1
    else: t[i]=0

#plot the simulated data
ax4 = axarr[1,1]
ax4.scatter(x, t)
ax4.axvline(0)
ax4.set_ylabel('class')
plt.show()

#compute the paratmer estimates using the synthetic training data
n2 = num_samples - n1
eGamma = n1 / float(num_samples)
zipped = zip(x,t)
eu1 = reduce(lambda accum, Z: accum + Z[0]*Z[1], zipped, 0) / float(n1)
eu2 = reduce(lambda accum, Z: accum + Z[0]*(1-Z[1]), zipped, 0) / float(n2)
def Sn(Z, ck, mu=(eu1,eu2)):
    idx = 1-int(ck)
    return (Z[0]-mu[idx])*(Z[0]-mu[idx])
eSigma = reduce(lambda accum, Z: accum + Sn(Z, Z[1]), zipped, 0) / float(num_samples)
print('The number of inputs in C1 is {0}'.format(n1))
print('The number of inputs that were assigned based on the assignment factor is {0}'.format(nae))
print('The true model parameters are gamma={0}, u1={1}, u2={2}, Sigma={3}'.format(tGamma, tu1, tu2, tSigma))
print('The estimated model parameters are gamma={0}, u1={1}, u2={2}, Sigma={3}'.format(eGamma, eu1, eu2, eSigma))
```

## Modèles discriminatoires probabilistes

<p style="text-align:justify">Les modèles discriminants probablistes supposent le même modèle linéaire généralisé
$ y (\phi (\mathbf{x})) = f (\mathbf{w}^T \phi (\mathbf{x})) $</p>

<p style="text-align:justify">comme les modèles génératifs probabilistes. Cependant, plutôt que de formuler des modèles pour les densités conditionnelles de classe, $ p (\phi (\mathbf{x}) | C_k) $, et les a priori de classe, $ p (C_k) $, l'approche discriminative modélise explicitement la classe postérieure probabilités, $ p (C_k | \phi (\mathbf {x})) $ avec les paramètres du modèle $ \mathbf{w} $. Comme dans l'approche générative probabiliste, le maximum de vraisemblance est utilisé pour estimer les paramètres du modèle à partir d'un ensemble de données d'apprentissage, $ \ mathbf {t} $. La différence est la forme de la fonction de vraisemblance. Dans le cas génératif probabiliste, la fonction de vraisemblance est une fonction de la probabilité conjointe, $ p (\phi (\mathbf {x}), C_k) = p (\phi (\mathbf {x}) | C_k) p (C_k ) $. Dans l'approche discriminante probabiliste, la fonction de vraisemblance est une fonction des postérieurs de la classe conditionnelle, $ p (C_k | \phi (\mathbf {x})) $ seulement.</p>

**REMARQUE** <p style="text-align:justify">La section sur les modèles génératifs probablistes s'est concentrée sur les modèles utilisant directement l'entrée $ \mathbf{x} $. Cependant, comme indiqué précédemment, ces modèles et les résultats sont également valables pour les entrées qui subissent une transformation fixe à l'aide d'un ensemble de fonctions de base, $ \phi $. Dans cette section, nous nous concentrerons sur l'inclusion d'une transformation d'entrée via les fonctions de base.</p>

### Logistic Regression

Nous considérons d'abord la régression logistique, qui est un exemple spécifique de modélisation discriminante, pour le cas de deux classes. Il suppose un modèle pour les probabilités postérieures de classe, $ p (C_k | \phi(\mathbf{x})) $, sous la forme du sigmoïde logistique

$ p (C_1 | \phi) = f(a) = \sigma (\mathbf{w}^T \phi) $

avec $ p (C_2 | \phi) = 1 - p (C_1 | \phi) $. Nous appliquons maintenant le maximum de vraisemblance pour obtenir les paramètres du modèle. Supposons que notre ensemble d'entraînement, $ \mathbf{t} $, soit de la forme $  {\phi_n,t_n } $ où $ \phi_n = \phi (\mathbf{x}_n) $, $ t_n \dans {0,1} $, et $ n = 1 \ldots N $. La fonction de vraisemblance des données d'apprentissage est alors

$ p (\mathbf{t} | \mathbf{w}) = \prod_{n = 1}^N \sigma (\mathbf{w}^T \phi (\mathbf{x}_n))^{t_n} (1 - \sigma (\mathbf{w}^T \phi (\mathbf{x}_n)))^{1-t_n} $

En définissant la fonction d'erreur, $ E(\mathbf{w}) $, comme le négatif de la fonction log-vraisemblance, et en prenant le gradient par rapport à $ \mathbf{w} $, nous obtenons

$ \bigtriangledown E(\mathbf {w}) = \sum_{n = 1}^N \left [\sigma (\mathbf{w}^T \phi (\mathbf{x}_n)) - t_n \right] \phi (\mathbf{x}_n) $

Superficiellement, cette fonction d'erreur ressemble à celle obtenue pour la régression linéaire sous l'hypothèse d'un modèle de bruit gaussien qui avait une solution de forme fermée. Cependant, la non-linéarité de la fonction sigmoïde, $ \sigma (\mathbf{w}^T \phi (\mathbf{x}_n)) $ empêche une solution de forme fermée dans le problème de régression logistique. Il faut donc appliquer une méthode itérative pour obtenir une solution numérique des paramètres, $ \mathbf{w} $. Nous allons ici considérer la méthode Newton-Raphson pour laquelle la minimisation de la fonction d'erreur prend la forme

$ \mathbf{w}^{\tau + 1} = \mathbf{w}^{\tau} - \mathbf{H}^{- 1} \bigtriangledown E(\mathbf{w}) $


où $ \mathbf{H} $ est la matrice de Hesse composée des secondes dérivées de la fonction d'erreur

$ \mathbf{H} = \bigtriangledown \bigtriangledown E (\mathbf{w}) = \Phi^T \mathbf{R} \Phi $


où $ \Phi $ est la matrice de conception $ N \times M $ dont $ n^{th} $ ligne est donnée par $ \phi (\mathbf{x_n}) ^T $ et $ \mathbf{R} $ est un $ N \times N $ matrice diagonale avec éléments $ R_{n, n} = \sigma (\mathbf{w}^T \phi (\mathbf{x}_n)) \left [1- \sigma (\mathbf{ w}^T \phi (\mathbf{x} _n)) \right] $. Cela peut être réduit à une forme équivalente à celle de la régression linéaire pondérée localement comme suit

$ \mathbf{w}^{\tau + 1} = \left (\Phi^T \mathbf{R} \Phi \right)^{- 1} \Phi^T \mathbf{R} \mathbf{z} $

où $ \mathbf{z} $ est un vecteur dimensionnel $ N $ défini par

$ \mathbf{z} = \Phi \mathbf{w}^{\tau} - \mathbf{R}^{- 1} (\mathbf{y} - \mathbf {t}) $

#### Exemple 2

Prenons un exemple avec deux classes et une entrée 2D, $ \mathbf{x_n} = (x_n^{(1)}, x_n^{(2)}) $. À titre expérimental, vous pouvez essayer d'augmenter le nombre de points d'entraînement, $ N $. Finalement, les points d'entraînement se chevaucheront de sorte qu'il ne sera pas possible de les séparer complètement avec la transformation fournie.

```python
from matplotlib import pyplot as plt

seed = 123456789
np.random.seed(seed)
N = 100 #number of data points, SEE WHAT HAPPENS AS YOU INCREASE THIS TO SAY 200
D = 2   #dimension of input vector
t = np.zeros(N) #training set classifications
X = np.zeros((N,D)) #training data in input space
sigma = .25
mu0 = 0.0
mu1 = 1.0

#func for generating 2D class scatter plot
def createScatter(X, t, ax):
    C1x = []  
    C1y = []
    C2x = []
    C2y = []
    for i in range(len(t)):
        if t[i] > 0: 
            C1x.append(X[i,0])
            C1y.append(X[i,1])
        else: 
            C2x.append(X[i,0])
            C2y.append(X[i,1])
    ax.scatter(C1x,C1y)
    ax.scatter(C2x,C2y, color='r')
    
#Generate test data. (NOTE THIS IS NOT BASED ON A GENERATIVE APPROACH. I AM PICKING SOMETHING THAT I KNOW (OK, THINK) WILL WORK. IN PRACTICE THIS DATA WOULD BE OBTAINED VIA OBSERVATION)
#Pick a value from a uniform distribution [0,1). If it is less than 0.5, assign class 1 and pick x1,x2 from a ND(mu0,sigma) otherwise assign class 2 and pick x1,x2 from ND(mu1,sigma)
for i in range(N):
    #choose class to sample for
    fac = 1
    if np.random.rand() <= 0.5:
        thismu = mu0
        t[i] = 1
    else: 
        thismu = mu1
        t[i] = 0
        if np.random.rand() < 0.5: fac = -1
        
    X[i,0] = fac * np.random.normal(thismu, sigma)
    X[i,1] = fac * np.random.normal(thismu, sigma)
    

f, axarr = plt.subplots(1,3)
f.subplots_adjust(right=2.5)
f.text(0.75,0.975,'Fixed Input Transformation Yields Linear Class Boundary',horizontalalignment='center',verticalalignment='top')

ax1 = axarr[0]
ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
createScatter(X,t,ax1)

#The training data does not have a linear boundary in the original input space. So lets apply a tranformation, phi to try to make it linearly seperable
#NOTE: This transformation is not the only one that works. For example try switching the values of MU1 and MU2. The result will be a different mapping
#that is still linearly seperable
def phi(x,mu,sigma):
    detSigma = np.linalg.det(sigma)
    fac = math.pow(2*math.pi, len(mu)/2.0) * math.sqrt(detSigma)
    arg = -0.5 * np.dot((x-mu).T, np.dot(np.linalg.inv(sigma), x-mu) )
    return math.exp(arg) / fac
    
phiX = np.zeros((N,D))
MU1 = np.ones(D)*mu0
MU2 = np.ones(D)*mu1
SIGMA = np.diag(np.ones(D))*sigma
for i in range(N):
    phiX[i,0] = phi(x=X[i,:], mu=MU2, sigma=SIGMA)
    phiX[i,1] = phi(x=X[i,:], mu=MU1, sigma=SIGMA)
    
ax2 = axarr[1]
ax2.set_xlabel('phi1(x)')
ax2.set_ylabel('phi2(x)')
createScatter(phiX, t, ax2)

#Now lets apply machine learning to determine the boundary. We will assume M = 3, i.e. that there are 3 free parameters that is w = [w0, w1, w2]^T and phi_n = [1, phiX[0], phiX[1]]
M = 3
Phi = np.ones((N,M))
Phi[:,1] = phiX[:,0]
Phi[:,2] = phiX[:,1]
w = np.zeros(M)
R = np.zeros((N,N))
y = np.zeros(N)

def sigmoid(a):
   return 1.0 / (1.0 + math.exp(-a))

def totalErr(y,t):
    e = 0.0
    for i in range(len(y)):
        if t[i] > 0:
            e += math.log(y[i])
        else:
            e += math.log(1.0 - y[i])
    return -e

#start Newton-Raphson. As a stopping criteria we will use a tolerance on the change in the error function and a max number of iterations
max_its = 100
tol = 1e-2
w0 = [w[0]] 
w1 = [w[1]]
w2 = [w[2]]
err = []
error_delta = 1 + tol
current_error = 0
idx = 0
while math.fabs(error_delta)>tol and idx < max_its:
    #update y & R
    for i in range(N): 
        zipped = zip(w, Phi[i,:])
        y[i] = sigmoid(reduce(lambda accum, Z: accum + Z[0]*Z[1], zipped, 0))
        R[i,i] = y[i] - y[i]*y[i]
    #update w
    z = np.dot(Phi,w) - np.dot(np.linalg.pinv(R),y-t)
    temp = np.linalg.pinv(np.dot(np.dot(Phi.T,R),Phi))
    temp2 = np.dot(np.dot(temp, Phi.T),R)
    w = np.dot(temp2, z)
    w0.append(w[0])
    w1.append(w[1])
    w2.append(w[2])
    idx += 1
    temp = totalErr(y,t)
    error_delta = current_error - temp
    current_error = temp
    err.append(error_delta)
print('The total number of iterations was {0}'.format(idx))
print('The total error was {0}'.format(current_error))
print('The final change in error was {0}'.format(error_delta))
print('The final parameters were {0}'.format(w))    
    
#our decision boundary is now formed by the line where sigma(a) = 0.5, i.e. where a = 0, which for this example is where phi2 = -(w1/w2)phi1, i.e. where w * phi = .5
bdryx = (0,1)
bdryy = (-(w[0]+w[1]*bdryx[0])/w[2], -(w[0]+w[1]*bdryx[1])/w[2])
ax2.plot(bdryx, bdryy)

ax3 = axarr[2]
ax3.plot(w0, color = 'blue')
ax3.plot(w1, color = 'red')
ax3.plot(w2, color = 'green')
ax3.plot(err, color = 'black')
ax3.legend(("w0","w1","w2", "error delta"), loc='upper left')
ax3.set_xlabel('Newton-Raphson Iteration')
```
